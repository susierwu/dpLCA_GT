```python
%matplotlib inline
#from __future__ import division
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import plotly.express as px
```

<h3> Dynamic LCI: tree case </h3> 


DNY_CO2 tool: https://ciraig.org/index.php/project/dynco2-dynamic-carbon-footprinter/ 
 
 
 

```python
def yearly_IRF_cal(dynLCI, DCF):
    result_IRF = [] 
    if (isinstance(dynLCI, pd.Series) and isinstance(DCF, pd.Series) ): 
        pass
    else: 
        dynLCI = pd.Series(dynLCI)
        DCF = pd.Series(DCF)
        print("Raw data has been changed to pd.Series format")    
        
    if (len(dynLCI) != len(DCF)): 
        print("re-check if LCI/CF with the same length (esp. for Year0)")
    else: 
        all_dny_lci = dynLCI
        all_rev_dcf  =  DCF[::-1] #reverse order of DCF  #DO not use DCF.sort_values() 
        
        for i in range(1, len(all_dny_lci)+1):  
            lci_n = all_dny_lci[0:i]
            dcf_n = all_rev_dcf[len(all_dny_lci)-i:len(all_dny_lci)]  
            result_IRF.append(sum(x * y for x, y in zip(lci_n, dcf_n )))
    return pd.Series(result_IRF)
```

<h3>READ in dynamic LCI </h3>

```python
LCI_dt = pd.read_excel('example_data/LCI_TREE_input.xlsx') 
LCI_dt
```

```python
LCI_101 = LCI_dt.loc[ :100, :]
```

### Read in DCF

```python
gas = "CH4"
MY = 2040
gas_dcf_list = [] 

dcf_single_sheet_name = "dcf_pointvalue" + gas + "_" + str(MY) 
agwp_single_sheet_name = "agwp_pointvalue" + gas + "_" + str(MY) 
f_dir = "/Users/susierwu/Dropbox/DLCA_DynCO2/pdpLCA_GT/FaIR_dpCFs/output/metrics/"  
scenarios = ["ssp119", "ssp126", "ssp245", "ssp585" ]

for scn in scenarios: 
    fname = (str("agwp_dcf_gwp100_tstep1" + gas + "_"+ scn + "_fair_start2000MY" + str(MY) + ".xlsx" ))
    file_path = os.path.join(f_dir, fname)
    df = pd.read_excel(file_path, sheet_name = dcf_single_sheet_name)
    gas_dcf_list.append(df)

finaldf = []
for i in range(len(scenarios)):
    colname = [ str(scenarios[i] ) ]
    dd = pd.DataFrame(zip(gas_dcf_list[i][gas]), columns = colname)
    finaldf.append(dd)
    
dcf = pd.concat(finaldf, axis=1 )

dcf
```

```python
df = pd.DataFrame(columns=scenarios)

for scn in scenarios:
    col_IRF = yearly_IRF_cal(LCI_101[gas], dcf[scn])
    df[scn] = col_IRF
    
df = df.iloc[1:] #delete first Y0

fig = px.line(df , title='Instantaneous Impacts AGFP [W/m2]_' + gas + "_MY" + str(MY))

fig.update_yaxes(exponentformat="e")  #https://plotly.com/python/reference/layout/yaxis/#layout-yaxis-autotypenumbers
fig.update_layout(
    yaxis_title="IRF[W m-2]",
    legend_title=gas
)
fig.show()
```
