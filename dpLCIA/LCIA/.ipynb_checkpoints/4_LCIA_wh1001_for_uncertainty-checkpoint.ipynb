### to check, mapping all minor GHGs,  or can method.write() seperately for each gas

```python
class assign_dpIRF:
     
    def __init__(self,
        cf_inputds, 
        # must be one of the value from rawCFs dataset 'dpCRF', 'dpGWP'
        ssp = '119',
        fairMY = 2030, 
        metric = 'dpIRF',  
        gas = 'CO2', #CH4, N2O
        majorGHG_names  = ['Carbon dioxide, non-fossil', 'Carbon dioxide, in air', 'Carbon dioxide, to soil or biomass stock', 
             'Carbon dioxide, from soil or biomass stock', 'Carbon dioxide, fossil', 'Carbon dioxide, non-fossil, resource correction', 
             'Methane, from soil or biomass stock', 'Methane, fossil', 'Methane, non-fossil' , 
              'Dinitrogen monoxide'],  
        ):
        
        """Initialise"""
        self.cf_inputds = cf_inputds
        self.ssp = ssp
        self.fairMY = fairMY
        self.metric = metric
        self.gas = gas
        self.majorGHG_names =  majorGHG_names 
    
    def prep_data_wh_uncert (self  ):
     
        data = []   
        TH = len(self.cf_inputds.Year) 
        for year in range(99, TH): # to loop through all years, each year, get the 1001 emsemble, get raw median (amount), mu (loc) and sd (scale)
            cf_t = []
            for stressor in self.majorGHG_names:
                if "Carbon dioxide" in stressor: 
                    gas_array_allY = self.cf_inputds.sel(SSP = self.ssp, dpLCIA = self.metric, ModelYear = self.fairMY, GHG = "CO2")['dpCF'].values  #wo Year, shape is (101, 1001) 
                    gas_array = gas_array_allY[year]
                    gas_amt_raw = np.median(gas_array)
                    gas_loc_raw, gas_scale_raw =  stats.norm.fit(gas_array) 
                   
                    all_efs = [i for i in mybio if i['name'] == stressor]
                    for ef in all_efs: 
                        #print(ef)
                        if "Carbon dioxide, in air" in ef['name'] or  "Carbon dioxide, non-fossil, resource correction" in ef['name'] or "Carbon dioxide, to soil or biomass stock" in ef['name']:
                            gas_amt_prm, gas_loc_prm, gas_scale_prm = gas_amt_raw  * (-1), gas_loc_raw * (-1), gas_scale_raw  
                        elif "Carbon dioxide, non-fossil" in ef['name'] or "Carbon dioxide, from soil or biomass stock" in ef['name'] or 'Carbon dioxide, fossil' in ef['name'] :   
                            gas_amt_prm, gas_loc_prm, gas_scale_prm = gas_amt_raw , gas_loc_raw, gas_scale_raw
                        else:  
                            print(f"some missing flow")
                            #gas_amt_prm, gas_loc_prm, gas_scale_prm = gas_amt_raw , gas_loc_raw, gas_scale_raw

                        if self.metric == 'dpGWP': 
                            co2_dict = gas_amt_prm
                        else: 
                            co2_dict = {
                                        'amount': gas_amt_prm,
                                        'uncertainty type': 3,  #normal 
                                        'loc': gas_loc_prm,
                                        'scale': gas_scale_prm ,
                                        'negative':  True if gas_amt_prm < 0 else False  
                            }
                        #print(f" raw CF for {stressor}: {gas_amt_raw},  {gas_scale_raw} ")
                        #print(f" after adjusted per premise_gwp: mu: {gas_loc_prm}, sd: {gas_scale_prm} ")
                        stressor_code = ef['code']
                        # we still relying on 'biosphere3', so name it as it is for LCIA 
                        cf_i = (('biosphere3', stressor_code),  co2_dict )  
                        cf_t.append(  cf_i )

                
                elif "Methane" in stressor: 
                    gas_array_allY = self.cf_inputds.sel(SSP = self.ssp, dpLCIA = self.metric, ModelYear = self.fairMY,  GHG = "CH4")['dpCF'].values   
                    gas_array = gas_array_allY[year]
                    gas_amt = np.median(gas_array)
                    gas_loc, gas_scale =  stats.norm.fit(gas_array) 
                    
                    all_efs = [i for i in mybio if i['name'] == stressor]
                    for ef in all_efs: 
                        gas_dict_wh_uncert = {
                                    'amount': gas_amt,
                                    'loc': gas_loc,
                                    'scale': gas_scale,
                                    'uncertainty type': 3,  #normal 
                                    'negative':  False  
                        }
    
                        print(f" CF with uncert for {stressor}: mu: {gas_loc}, sd: {gas_scale} ")
                        
                        stressor_code = ef['code']
                        # we still relying on 'biosphere3', so name it as it is for LCIA 
                        cf_i = (('biosphere3', stressor_code),  gas_dict_wh_uncert )  
                        cf_t.append(  cf_i )
                        
                     
                elif "Dinitrogen monoxide" in stressor: 
                    gas_array_allY = self.cf_inputds.sel(SSP = self.ssp, dpLCIA = self.metric, ModelYear = self.fairMY, GHG = "N2O")['dpCF'].values  
                    gas_array = gas_array_allY[year]
                    gas_amt = np.median(gas_array)
                    gas_loc, gas_scale =  stats.norm.fit(gas_array) 
                    
                    all_efs = [i for i in mybio if i['name'] == stressor]
                    for ef in all_efs: 
                        gas_dict_wh_uncert = {
                                    'amount': gas_amt,
                                    'loc': gas_loc,
                                    'scale': gas_scale ,
                                    'uncertainty type': 3,  #normal 
                                    'negative':   False  
                        }
                        print(f" CF with uncert for {stressor}: mu: {gas_loc}, sd: {gas_scale} ")

                        stressor_code = ef['code']
                        # we still relying on 'biosphere3', so name it as it is for LCIA 
                        cf_i = (('biosphere3', stressor_code),  gas_dict_wh_uncert )  
                        cf_t.append(  cf_i )
                    
            data.append(cf_t)
 
        return data 


    def prep_final_dpCFs_wh_uncert (self, data):
            """ 
            write BW2method from data 
            Inputs: 
                data from prep_data_wh_uncert()
            Returns: 
                null, write to BW2 project directly 
            """ 
            ssp, MY = str('SSP' + self.ssp) , str('MY' + str(self.fairMY))   
            m_name = 'IPCC 2021 - dpCFs_wh_uncert - ' + ssp + '_' + MY + ' - year'
            print(f" start creating new method, with name: {m_name}")
            
            for i in range( 0, len(data) ):   
                name = (m_name + str(i) ,  'climate change' , self.metric + str(i) ) 
                new_method = bw2data.Method(name)
                new_method.register()
                if self.metric == 'dpIRF': 
                    new_method.metadata["unit"] = 'W m-2 kg-1'
                elif self.metric == 'dpCRF': 
                    new_method.metadata["unit"] = 'W m–2 yr kg–1'
                elif self.metric == 'dpGWP': 
                    new_method.metadata["unit"] = 'kgCO2e'
                new_method.write(data[i])
            print("finishing preparing new methods")
```
