```python
import bw2data, bw2io
import bw2calc
import os
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colormaps
import numpy as np
import pandas as pd
import xarray as xr
plt.rcParams['font.family'] = 'Times New Roman'
matplotlib.style.use('tableau-colorblind10') #seaborn-v0_8-whitegrid  tableau-colorblind10
```

### prepare empty DF under SSPx, MY2020/2030/2040/2050

```python
len_cf = 101  # add 1 as CFs value starting from year_0
MYs = [2020, 2030, 2040, 2050] #, 2060, 2070]
ssps = ["119", "245", "585"]   #"126"
sp_col, my_col = [], []

for sp in ssps:
    repeat_sp = [sp] * len_cf * len(MYs)
    sp_col.append(repeat_sp)
sp_col_f = my_col_f = [item for sublist in sp_col for item in sublist]

for my in MYs: 
    repeat_my = [my] * len_cf  
    my_col.append(repeat_my)

my_col = my_col * len(ssps)
my_col_f = [item for sublist in my_col for item in sublist]

print(len(sp_col_f), len(my_col_f))
```

```python
year_col =  [np.arange(len_cf)] *len(ssps) * len(MYs)
year_col_f =  [item for sublist in year_col for item in sublist]
print(len(year_col_f))
```

### read dpLCIA metrics prepared under SSPx, MY2020/2030/2040/2050

```python
ghggass = ["CO2", "CH4", "N2O"]
cf_col_gas_allgas = []

for ghggas in ghggass: 
    cf_col_gas = []
    for sp in ssps:
        for MY in MYs:
            #print(f"read point GWP value for {ghggas} SSP {sp}, model year {MY}")
            filename = "agwp_dcf_gwp100_tstep1" + ghggas + "_ssp" + sp + "_fair_start2000MY" + str(MY) + ".xlsx"
            cf_file = os.path.join("/Users/susierwu/Dropbox/DLCA_DynCO2/pdpLCA_GT/FaIR_dpCFs/output/metrics/", filename)
            cf_gwp_df = pd.read_excel(cf_file, sheet_name = "gwp_pointvalue" + ghggas+ "_" + str(MY) ) 
            cf_gwp_pointvalue = cf_gwp_df[ghggas].values
            cf_col_gas.append(cf_gwp_pointvalue)
    cf_col_gas_allgas.append(cf_col_gas)
```

```python
cf_col_gas_f = [item for sublist in cf_col_gas for item in sublist]
print(len(cf_col_gas_f))
```

```python
#co2_irf_all, ch4_irf_all, n2o_irf_all = cf_col_gas_allgas[0], cf_col_gas_allgas[1], cf_col_gas_allgas[2] 
gas_all_f_list = []

# gas_irf_all_f_list[i] is the ith gas from ghggass, flattened array for all SSPs, MYt in order 
for i in range(len(cf_col_gas_allgas)): 
    gas_all_f = [item for sublist in cf_col_gas_allgas[i] for item in sublist]
    gas_all_f_list.append(gas_all_f)
```

```python
cf_df = pd.DataFrame(zip(sp_col_f, my_col_f, year_col_f, gas_all_f_list[0], gas_all_f_list[1], gas_all_f_list[2]) , 
                     columns = ["SSP", "ModelYear", "Year", "CO2_GWP", "CH4_GWP", "N2O_GWP" ] ) 
```

```python
cf_df.set_index(['SSP', 'ModelYear', 'Year'], inplace=True)
```

### prepare and save final CF as xarray ds

```python
#Deleting rows with YEAR_index value 0
cf_df_filtered = cf_df[~cf_df.index.get_level_values("Year").isin([0])]
#cf_df_filtered
```

```python
cf_xr = cf_df.to_xarray() 
cf_xr2 = cf_df_filtered.to_xarray() 
```

```python
cf_xr.to_netcdf('data/CF_GWP0_100_perSSP_MY_majorghgs.nc')
cf_xr2.to_netcdf('data/CF_GWP1_100_perSSP_MY_majorghgs.nc')
```

### plotting dGWP20/100 grouped by MY

```python
ch4_gwp20, ch4_gwp100, n2o_gwp20, n2o_gwp100 = [] , [], [] , []
for MY in [2030, 2040, 2050]:  #2020
    for sp in ssps: 
        #for year in [20, 100]: 
        ch4 = cf_xr.sel(ModelYear = MY, Year = [20, 100], SSP = sp)['CH4_GWP']
        n2o = cf_xr.sel(ModelYear = MY, Year = [20, 100], SSP = sp)['N2O_GWP']
        #print(ch4.values)
        ch4_gwp20.append(ch4.values[0])
        ch4_gwp100.append(ch4.values[1]) 
        n2o_gwp20.append(n2o.values[0])
        n2o_gwp100.append(n2o.values[1])
```

```python
# prepare columns:
gas = ['CH4'] * 4 * len(ssps)
MY =  ['2030'] * 3 + ['2040'] * 3  + ['2050'] * 3  
SSPx = ["119", "245", "585"] * 4 
```

```python
ch420_100 = pd.DataFrame(zip(gas,MY,SSPx,ch4_gwp20,ch4_gwp100 ), columns = ["gas", "MY", "SSP", "GWP20", "GWP100"]) 
```

```python
n2o20_100 = pd.DataFrame(zip(gas,MY,SSPx,n2o_gwp20,n2o_gwp100 ), columns = ["gas", "MY", "SSP", "GWP20", "GWP100"]) 
```
