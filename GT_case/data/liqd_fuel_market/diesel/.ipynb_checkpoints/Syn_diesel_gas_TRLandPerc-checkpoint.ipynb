Synthetic Diesel:
Feedstock: Synthetic diesel is produced from fossil-based feedstocks such as natural gas, coal, or biomass through a process known as Fischer-Tropsch synthesis.

Biodiesel: Production Process: Biodiesel is produced through a process called transesterification, which involves reacting vegetable oils or animal fats with an alcohol (such as methanol or ethanol) and a catalyst to produce fatty acid methyl esters (FAME) or fatty acid ethyl esters (FAEE).

Renewable Diesel: Production Process: Renewable diesel is produced through a process called hydrotreating or hydroprocessing, which involves hydrogenation of triglycerides (found in vegetable oils and animal fats) and other renewable feedstocks to remove oxygen and saturate unsaturated bonds, resulting in a product that is chemically similar to petroleum diesel.


In summary, synthetic diesel, renewable diesel, and biodiesel are all alternative fuels that offer various environmental and energy security benefits compared to petroleum diesel. While synthetic diesel is produced from fossil-based feedstocks, renewable diesel and biodiesel are derived from renewable feedstocks, with renewable diesel being chemically similar to petroleum diesel and biodiesel typically used as a blend with petroleum diesel.
