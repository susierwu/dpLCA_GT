## 1. prepare empty DF  

### 1.A prepare empty DF 

```python
""" same EFs for different SSP """ 
v_year = [2030, 2040, 2050]
v_run_y = list(range(1, 21))
v_list = ['garbage_truck' ]
size_GVWR = ['HDV_class7']
powertrain = ['ICEV-d',  'BEV' ]
cc_pollutants = ['exh_CO2', 'exh_CH4', 'exh_N2O']
```

```python
list_pollut = [] 
for pp in cc_pollutants: 
    list_pollut.append( [pp] * len(v_run_y))
#list_pollut

list_year = [] 
for yr in v_year: 
    list_year.append( [yr] *  (len(list_pollut[0]) * len(list_pollut)) )

list_pt = []
for pt in powertrain: 
    list_pt.append( [pt] * (len(list_year[0]) * len(list_year))  )

list_size = []
for sz in size_GVWR: 
    list_size.append( [sz] * (len(list_pt[0]) * len(list_pt))  )


list_vh = []
for vt in v_list: 
    list_vh.append([vt] * len(list_size[0]) * len(list_size) )
 
```

```python
print( (len(list_vh[0]) * len(list_vh)) ,  (len(list_size[0]) * len(list_size)),  (len(list_pt[0]) * len(list_pt)),  (len(list_year[0]) * len(list_year)) ,  (len(list_pollut[0]) * len(list_pollut))  )
```

```python
list_size_f = [item for sublist in list_size for item in sublist]  
len_size = int((len(list_vh[0]) * len(list_vh)) / (len(list_size[0]) * len(list_size))) 
list_size2 = [list_size_f] * len_size
len(list_size2[0]) * len(list_size2) == (len(list_vh[0]) * len(list_vh))
```

```python
list_pt_f = [item for sublist in list_pt for item in sublist]  
len_pt = len_size *  int((len(list_size[0]) * len(list_size)) / (len(list_pt[0]) * len(list_pt))) 
list_pt2 = [list_pt_f] * len_pt
len(list_pt2[0]) * len(list_pt2) == (len(list_vh[0]) * len(list_vh))
```

```python
list_yr_f = [item for sublist in list_year for item in sublist]  
len_yr = len_pt *  int((len(list_pt[0]) * len(list_pt)) / (len(list_year[0]) * len(list_year)))
list_yr2 = [list_yr_f] *  len_yr
len(list_yr2[0]) * len(list_yr2) == (len(list_vh[0]) * len(list_vh))
```

```python
list_pollut_f = [item for sublist in list_pollut for item in sublist]  
len_pollut = len_yr * int( (len(list_year[0]) * len(list_year)) / ( len(list_pollut) * len( v_run_y )) )
len_pollut2 = [list_pollut_f] *   len_pollut 
len(len_pollut2[0]) * len(len_pollut2) == (len(list_vh[0]) * len(list_vh))
```

```python
def flatten_list(lst):
    return [item for sublist in lst for item in (flatten_list(sublist) if isinstance(sublist, list) else [sublist])]
```

```python
pollutf, yrf, ptf, sizef, vhf =  flatten_list(len_pollut2), flatten_list(list_yr2), flatten_list(list_pt2),flatten_list(list_size2),  flatten_list(list_vh)
```

### 1.B prepare DF. adding empty cols  

```python
finaldf = pd.DataFrame(zip(vhf, sizef, ptf, yrf, pollutf), columns=['v_type', 'size', 'powertrain', 'year', 'pollutants'])
```

```python
# add run_Y col
v_run_y_col = v_run_y * int(len(finaldf) / len(v_run_y))
finaldf["running_y"] = v_run_y_col
finaldf["emission_kgperkgd"] = np.full(int(len(finaldf)), np.nan)
finaldf["emission_kgperyear"] = np.full(int(len(finaldf)), np.nan)
```

```python
finaldf['diesel_kgperyear'] =  np.full(int(len(finaldf)), np.nan)
finaldf
```

## 2. get diesel fuel consumption - MOVES EF doens't give us correct CO2/CH4/N2O emissions, using NRCAN EFs 

```python
root_path = '/Users/susierwu/Dropbox/DLCA_DynCO2/pdpLCA_GT36/'
consmp_p_file = os.path.abspath(os.path.join(root_path, 'data/P_files/tech_conversion_veh_p/p_scn_v_need_tech.xlsx'))
v_lifetime = 20 
v_yearlyFU = [36000] * v_lifetime  #assuming each year running same mileage

for year_to_get_inP in [2030, 2040, 2050]:
    for ssp_to_get_inP in ["ssp119", "spp126",  "spp245", "spp585"]: 
    
        v_pneed = loadYearlyParamswhsize(PARAMS_FILE = consmp_p_file, s_name = 'g_truck', hypothesis = 'point value', 
                                   years = [year_to_get_inP], SSP = ssp_to_get_inP, size = 'HDV_class7')
    
        for Y in range(1, v_lifetime + 1,  1):  
    
            regex_pattern = re.compile("diesel_consump_per") # it could be per km or per hour
            for key in v_pneed:
                if regex_pattern.search(key): 
                    d_consumperkm = v_pneed[key] 
                    break
                else:
                    d_consumperkm = 0.42 # kg per km a static default value
    
            d_consumperkm2 = d_consumperkm[0] if isinstance(d_consumperkm, list) else d_consumperkm  
            tt = (Y + year_to_get_inP ) - year_to_get_inP - 1 # [tt] is the list_index if v_yearlyFU is a list, -1 if it starts running from Y[1], not Y[0]
            d_consumperyear = v_yearlyFU[tt]  * d_consumperkm2
            print(f"  ⛽️⛽️ for v_year{year_to_get_inP}, running_year {Y}, SSP {ssp_to_get_inP}, the annual diesel_consump is {d_consumperyear}")  

            """ assign the annual diesel consumption to DF, only to ICEV-d, BEV has 0 CC emissions  """ 
            sel_indx = finaldf.index[(finaldf["year"] == year_to_get_inP) & (finaldf["running_y"] == Y) & (finaldf["powertrain"] == "ICEV-d" )].tolist()
            finaldf.loc[sel_indx, 'diesel_kgperyear'] =  d_consumperyear
        
```

### 2. GET CO2 / CH4 / N2O from NRCAN 

```python
""" 
from NRCAN: 
Heavy-duty Diesel Vehicles (HDDVs)	Emission Factors (g/L fuel)	CO2 CH4 N2O 		
Advanced Control	2,680.50	a	0.11	i	0.151
Moderate Control	2,680.50	a	0.14	f	0.082

assuming 0.84	kg per liter diesel, regardless if bio-diesel or conventional diesel, as for the F2_fuel_diesel module 
https://docs.google.com/spreadsheets/d/1_qotcARay-UyygXMX8Ha16aR90uKgEH81npOohgFgcg/edit#gid=654492837 
""" 

# assuming for v_year 2020 model::  Y0 are the Moderate Control, Y10, Y20 all same yearly value 
# for v_year after 2030, Y0 using the Advanced Control 
""" prepare a dict, key value same as in the DF, unit is kgperkgd """

v2030_ef = {
    'exh_CO2': 2.68 / 0.84  ,
    'exh_CH4': 0.11 / 1000 / 0.84 ,
    'exh_N2O': 0.082/ 1000 / 0.84 
}
```

```python
### regardless of running_y, assign same EFs

for vyear in  [ 2030, 2040, 2050]:
    for pol in ["exh_CO2", "exh_CH4", "exh_N2O"]: 

        sel_ind = finaldf.index[(finaldf["year"] == vyear) & (finaldf["powertrain"] == "ICEV-d" ) & (finaldf["pollutants"] == pol) ]
        finaldf.loc[sel_ind, 'emission_kgperkgd'] =  v2030_ef.get(pol)
```

```python
finaldf["emission_kgperyear"] = finaldf["emission_kgperkgd"] * finaldf["diesel_kgperyear"]
```
