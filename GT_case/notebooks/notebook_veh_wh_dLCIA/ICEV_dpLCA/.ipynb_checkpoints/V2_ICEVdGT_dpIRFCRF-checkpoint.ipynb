### A1. prepare empty ds for assigning dLCIA score once its prepared, same ds as in other modules

```python
v_lifetime = 20 
# Define dimensions
dpLCA_type = ['dpIRF', 'dpCRF', 'dpGWP']
SSP = ['ssp119', 'ssp245', 'ssp585']
v_year = [2030,2040,2050]
tech_list = ['dummy_tech']
v_module = ['V2_oper_EF']

""" running year starts from Y[1], not incl. initial manuf. YEAR[0], 100 is the LCIA_TH here, dpIRF has total 101 values,
    so for running_year[1] emission, TOD is 1 - 101 (with 101 CFs, starting right from the pulse emission at year[1]), 
    for running_year[20] emission, TOD is 20 - 120 (with 101 CFs, starting from year[20])
"""
lcia_year = np.arange(1, 101+v_lifetime, 1) 
ensems = np.arange(1001) 
# Create a new xarray dataset with dimensions
cord = {'SSP': SSP, 'v_year': v_year, 'v_module': v_module, 'tech_list':tech_list, 'lcia_year': lcia_year, 'dpLCA_IC': dpLCA_type, 'ensemble': ensems}
ds = xr.Dataset(coords=cord)

#   data array with fake data
fakedd = np.full((len(SSP), len(v_year), len(v_module), len(tech_list), len(lcia_year), len(dpLCA_type), len(ensems) ), np.nan)    
dplcia = xr.DataArray(fakedd, dims=('SSP', 'v_year', 'v_module', 'tech_list', 'lcia_year', 'dpLCA_IC', 'ensemble'), coords=cord )
 
# Assign the data variable to the dataset
ds = ds.assign(dplcia = dplcia)
ds
```

## C. calc. prep everything per SSP[x], v_year[t] and overwrite ds directly in the loop

#### C1. testing code

```python
for sp in ['119', '245', '585']:  #caz cf_xr now have SSP defined as 119/126/245/585, to be re-do: LCIA/2_CFanal_IRF_fairSY2000.ipynb 
    for v_y  in [2030, 2040, 2050]:
        v_year = v_y
        sp_toget = sp
        finaldf_list = [] 
        for gas in  ['CO2', 'CH4', 'N2O']: 
            DCF_per10years = []
            for my in MYs:
                newcf_per10yr = cf_xr.sel(SSP = sp_toget, ModelYear = v_y)[gas+"_IRF"].values
                DCF_per10years.append(newcf_per10yr)
            
            #print(len(DCF_per10years))
            if v_year == 2030: 
                vy_i = 0
            elif v_year ==2040: 
                vy_i = 1
            elif v_year == 2050: 
                vy_i = 2

            all_lci_irf_y_list = []
            #all_dny_lci: the annual emissions for v_lifetime 
            all_dny_lci  =  LCI_dt_allY[vy_i][gas][:v_lifetime] 
            raw_irf = DCF_per10years[vy_i][0:101]  
            
            
            #print(len(all_dny_lci), len(raw_irf))
            
            for yy in range(0, len(all_dny_lci)  ):  #len(dny_lci) #year 0 is 0, starting year 1 
                dny_lci = all_dny_lci[  : yy+1 ] 
                print(f" annual emission of {gas} for running_year {len(dny_lci.values )} is {dny_lci.values[yy] } ") 
                 
                """ TH up till last emission + 101, the same 101 CF multipled to year[yy]_lci, if need fixed TH, 
                    then use a CF list for each Year[yy]_lci :  (len(raw_irf[:101 - yy])) """ 
                all100impact_for_year_yy_lci = [x * dny_lci.values[yy] for x in raw_irf ]  
                #print(len(all100impact_for_year_yy_lci))
                #print(all100impact_for_year_yy_lci)
                all_lci_irf_y_list.append(all100impact_for_year_yy_lci) 
            #print(len(all_lci_irf_y_list) , len(all_lci_irf_y_list[0]), len(all_lci_irf_y_list[1]))
```

### direct calculation for V2, no BW2

```python
for sp in ['119', '245', '585']:  #caz cf_xr now have SSP defined as 119/126/245/585, to be re-do: LCIA/2_CFanal_IRF_fairSY2000.ipynb 
    for v_y  in [2030, 2040, 2050]:
        v_year = v_y
        sp_toget = sp
        finaldf_list = [] 
        for gas in  ['CO2', 'CH4', 'N2O']: 
            DCF_per10years = []
            for my in MYs:
                newcf_per10yr = cf_xr.sel(SSP = sp_toget, ModelYear = v_y)[gas+"_IRF"].values
                DCF_per10years.append(newcf_per10yr)
            
            #print(len(DCF_per10years))
            if v_year == 2030: 
                vy_i = 0
            elif v_year ==2040: 
                vy_i = 1
            elif v_year == 2050: 
                vy_i = 2

            all_lci_irf_y_list = []
            #all_dny_lci: the annual emissions for 20 years v_lifetime 
            all_dny_lci  =  LCI_dt_allY[vy_i][gas][:v_lifetime] 
            raw_irf = DCF_per10years[vy_i][0:101]  
            
            for yy in range(0, len(all_dny_lci)  ):  #len(dny_lci) #year 0 is 0, starting year 1 
                dny_lci = all_dny_lci[  : yy+1 ] 
                #print(f" annual emission of {gas} for running_year {len(dny_lci.values )} is {dny_lci.values[yy] } ") 
                """ TH up till last emission + 101, the same 101 CF multipled to year[yy]_lci, if need fixed TH, 
                    then use a CF list for each Year[yy]_lci :  (len(raw_irf[:101 - yy])) """ 
                all100impact_for_year_yy_lci = [x * dny_lci.values[yy] for x in raw_irf ]  
                #print(len(all100impact_for_year_yy_lci))
                #print(all100impact_for_year_yy_lci)
                all_lci_irf_y_list.append(all100impact_for_year_yy_lci) 
            #print(len(all_lci_irf_y_list) , len(all_lci_irf_y_list[0]), len(all_lci_irf_y_list[1]))

            """ prepare final DF for each gas,  summed together """ 
            all_col = [] #need 20 columns here, each col
            for t in range(len(all_lci_irf_y_list)):
                each_y_col = [0] * t + all_lci_irf_y_list[t]
                all_col.append(each_y_col)
            #print(len(all_col), len(each_y_col))
            max_length = len(each_y_col)
            # Create a dictionary where keys are column names and values are lists padded with np.nan
            data_dict = {}
            for i in range(max_length):
                data_dict[f'Year_{i+1}'] = [sublist[i] if i < len(sublist) else np.nan for sublist in all_col]
            # Create a DataFrame from the dictionary
            df = pd.DataFrame(data_dict)
            finaldf_gasx = df.T
            finaldf_list.append(finaldf_gasx)
        #print(f"len of finaldf_list: {len(finaldf_list)}") # it's always three then we add them together for CO2/CH4/N2O
        finalallghgdf = finaldf_list[0] + finaldf_list[1]  + finaldf_list[2]  
        ds_lciadata  = finalallghgdf.sum(axis=1)
        
        irf_score = np.array(ds_lciadata)
        crf_score_bad = np.cumsum(irf_score)
        # using cumtrapz for time-integral, more accurate to calculate CRF, slightly different from cumsum above 
        xt = np.arange(irf_score.shape[0])              
        crf_score = cumtrapz(irf_score, xt, initial=0)
        #print(crf_score_bad, "\n", crf_score)    
        print( f" calculated dpIRF and dpCRF score shape is: {irf_score.shape}, {crf_score.shape} ") 
        
        sp_in_ds =  'ssp' + str(sp) 
        to_addvar_ds_shape = ds['dplcia'].loc[{'SSP': sp_in_ds, 'v_year': v_y, 'v_module':'V2_oper_EF', 'dpLCA_IC': 'dpIRF', 'tech_list':'dummy_tech', 'ensemble':0}].shape
        if to_addvar_ds_shape == irf_score.shape == crf_score.shape: 
            for impact in  ['dpIRF', 'dpCRF', 'dpGWP']: 
                if impact == "dpIRF": 
                    ds['dplcia'].loc[{'SSP': sp_in_ds, 'v_year': v_y, 'v_module':'V2_oper_EF', 'tech_list':'dummy_tech', 'dpLCA_IC':impact, 'ensemble':0 }]  = irf_score
                elif impact == 'dpCRF': 
                    ds['dplcia'].loc[{'SSP': sp_in_ds, 'v_year': v_y, 'v_module':'V2_oper_EF', 'tech_list':'dummy_tech', 'dpLCA_IC':impact, 'ensemble':0 }]  = crf_score
                else: 
                    pass
                    #print("no dpLCA score for dpGWP yet, only absolute metrics prepared for dpLCIA")
            
        else:
            print(f"wrong ds_dim for SSP {sp}, year {v_y}")
```
