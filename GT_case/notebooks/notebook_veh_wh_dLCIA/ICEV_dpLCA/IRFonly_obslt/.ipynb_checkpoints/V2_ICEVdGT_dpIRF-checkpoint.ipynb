## B. test with each gas, TH up till last emission + 100  with: <font color = 'red'> a.  MY2020 IRF for v_year2020, b. MY2030 IRF for v_year2030, c. MY2040 IRF for v_year2040 ... </font>

#### B.1 CO2 

```python
v_year = 2030
sp_toget = '119'
gas='CO2'

DCF_per10years = []
for my in MYs:
    newcf_per10yr = cf_xr.sel(SSP = "119", ModelYear = my)[gas+"_IRF"].values
    DCF_per10years.append(newcf_per10yr)

print(len(DCF_per10years))


if v_year == 2030: 
    vy_i = 0
elif v_year ==2040: 
    vy_i = 1
elif v_year == 2050: 
    vy_i = 2
```

```python
all_lci_irf_y_list = []
all_dny_lci  =  LCI_dt_allY[vy_i][gas][:20] 
raw_irf = DCF_per10years[vy_i][0:100]  

for yy in range(0, len(all_dny_lci)  ):  #len(dny_lci) #year 0 is 0, starting year 1 
    dny_lci = all_dny_lci[  : yy+1 ] 
    #print(dny_lci.values[yy] ) 
    """ TH up till last emission + 100, the same 100 CF multipled to year[yy]_lci, if need fixed TH, 
        then use a CF list for each Year[yy]_lci :  (len(raw_irf[:100 - yy])) """ 
    all100impact_for_year_yy_lci = [x * dny_lci.values[yy] for x in raw_irf ] #raw_irf[:100 - yy]
    #print(len(all100impact_for_year_yy_lci))
    #print(all100impact_for_year_yy_lci)
    all_lci_irf_y_list.append(all100impact_for_year_yy_lci) 


print(len(all_lci_irf_y_list) , len(all_lci_irf_y_list[0]), len(all_lci_irf_y_list[1]))
```

##### plot dIRF for gas

```python
# Initialize x-axis counter
x_offset = 0

# Plot each sublist with x-axis shifted by 10 units for each new sublist
for sublist in all_lci_irf_y_list:
    x_values = [x + x_offset for x in range(len(sublist))]
    plt.plot(x_values, sublist)  #, marker='o' , label=f'Sublist {all_lci_irf_y_list.index(sublist) + 1}'
    x_offset += 1  # Increase x_offset for the next sublist

# Add labels and legend
plt.xlabel('Year')
plt.ylabel('W/m2')
plt.title( gas + '_IRF from 1st year with Shifted Year')
plt.legend()
```

#### prepare a final DF for summing

```python
all_col = [] #need 20 columns here, each col
for t in range(len(all_lci_irf_y_list)):
    each_y_col = [0] * t + all_lci_irf_y_list[t]
    all_col.append(each_y_col)

print(len(all_col), len(each_y_col))

max_length = len(each_y_col)
# Create a dictionary where keys are column names and values are lists padded with np.nan
data_dict = {}
for i in range(max_length):
    data_dict[f'Year_{i+1}'] = [sublist[i] if i < len(sublist) else np.nan for sublist in all_col]

# Create a DataFrame from the dictionary
df = pd.DataFrame(data_dict)

finaldf_CO2 = df.T
finaldf_CO2
```

```python
finaldf_CO2.sum(axis=1).plot( title = gas + '_dIRF [W/m2] for each year')
```

```python
outf_name =  "ICEV-d_MY" + str(v_year) + "_SSP" + str(sp_toget) + gas +  "_IRF_THtolci_y" + '.xlsx'
finaldf_CO2.to_excel(os.path.join("Interm_EFs_results/", outf_name ))
```

## C. calc. prep everything per SSP[x], v_year[t] and overwrite ds directly in the loop

```python
for sp in ['119', '126', '245', '585']:
    for v_y  in [2030, 2040, 2050]:

        v_year = v_y
        sp_toget = sp

        finaldf_list = [] 
        for gas in  ['CO2', 'CH4', 'N2O']: 
        
        
            DCF_per10years = []
            for my in MYs:
                newcf_per10yr = cf_xr.sel(SSP = sp_toget, ModelYear = v_y)[gas+"_IRF"].values
                DCF_per10years.append(newcf_per10yr)
            
            print(len(DCF_per10years))

            if v_year == 2030: 
                vy_i = 0
            elif v_year ==2040: 
                vy_i = 1
            elif v_year == 2050: 
                vy_i = 2


            all_lci_irf_y_list = []
            all_dny_lci  =  LCI_dt_allY[vy_i][gas][:20] 
            raw_irf = DCF_per10years[vy_i][0:100]  
            
            for yy in range(0, len(all_dny_lci)  ):  #len(dny_lci) #year 0 is 0, starting year 1 
                dny_lci = all_dny_lci[  : yy+1 ] 
                #print(dny_lci.values[yy] ) 
                """ TH up till last emission + 100, the same 100 CF multipled to year[yy]_lci, if need fixed TH, 
                    then use a CF list for each Year[yy]_lci :  (len(raw_irf[:100 - yy])) """ 
                all100impact_for_year_yy_lci = [x * dny_lci.values[yy] for x in raw_irf ] #raw_irf[:100 - yy]
                #print(len(all100impact_for_year_yy_lci))
                #print(all100impact_for_year_yy_lci)
                all_lci_irf_y_list.append(all100impact_for_year_yy_lci) 

            #print(len(all_lci_irf_y_list) , len(all_lci_irf_y_list[0]), len(all_lci_irf_y_list[1]))

            """ prepare final DF for each gas, will be summed together """ 
            
            all_col = [] #need 20 columns here, each col
            for t in range(len(all_lci_irf_y_list)):
                each_y_col = [0] * t + all_lci_irf_y_list[t]
                all_col.append(each_y_col)
            
            print(len(all_col), len(each_y_col))
            
            max_length = len(each_y_col)
            # Create a dictionary where keys are column names and values are lists padded with np.nan
            data_dict = {}
            for i in range(max_length):
                data_dict[f'Year_{i+1}'] = [sublist[i] if i < len(sublist) else np.nan for sublist in all_col]
            
            # Create a DataFrame from the dictionary
            df = pd.DataFrame(data_dict)
            finaldf_gasx = df.T
            finaldf_list.append(finaldf_gasx)

        print(f"len of finaldf_list: {len(finaldf_list)}")
        finalallghgdf = finaldf_list[0] + finaldf_list[1]  + finaldf_list[2]  

        ds_lciadata =  finalallghgdf.sum(axis=1)

        print( f" calculated lcia score shape is: {np.array(ds_lciadata).shape} ") 

        sp_in_ds =  'ssp' + str(sp) 
        if np.array(ds_lciadata).shape ==  ds['dplcia'].loc[{'SSP':sp_in_ds , 'v_year': v_y, 'v_module':'V2_oper_EF', 'tech_list':'dummy_tech', 'ensemble':0}].shape: 
            ds['dplcia'].loc[{'SSP': sp_in_ds, 'v_year': v_y, 'v_module':'V2_oper_EF', 'tech_list':'dummy_tech', 'ensemble':0}] = np.array(ds_lciadata)
        else:
            print(f"wrong ds_dim for SSP {sp}, year {v_y}")
```
