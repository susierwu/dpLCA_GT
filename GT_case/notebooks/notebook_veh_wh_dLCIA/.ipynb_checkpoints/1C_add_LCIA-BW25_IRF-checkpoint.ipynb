#### all EFs for majorGHGs

```python
majorGHG_names = ['Carbon dioxide, non-fossil', 'Carbon dioxide, in air', 'Carbon dioxide, to soil or biomass stock', 
             'Carbon dioxide, from soil or biomass stock', 'Carbon dioxide, fossil', 'Carbon dioxide, non-fossil, resource correction', 
             'Methane, from soil or biomass stock', 'Methane, fossil', 'Methane, non-fossil' , 
             'Dinitrogen monoxide']
```

#### Prepare a new df - C as CF data for the new LCIA

- https://github.com/polca/premise_gwp/tree/main
"Carbon dioxide, in air", with a CF of -1
"Carbon dioxide, non-fossil, resource correction", with a CF of -1
"Carbon dioxide, non-fossil", with a CF of +1

```python
indexrow = []
for i in range(101): 
    x36 = "IRF"+str(i)
    indexrow.append(x36)
newcol = majorGHG_names

C = pd.DataFrame(0, index=indexrow,columns = newcol)
C.head()
```

```python
"""
let's start with SSP119_MY2020
"""
cf_touse = cf_point.sel(SSP='119', ModelYear=2020)


for col in C.columns:
    if "Carbon dioxide" in col: 
        if "Carbon dioxide, in air" in col:
            #print(col) 
            C[col] = cf_touse['CO2_IRF'].values * (-1)
        elif "Carbon dioxide, non-fossil, resource correction" in col:
            C[col] = cf_touse['CO2_IRF'].values * (-1)
        elif "Carbon dioxide, non-fossil" in col or 'Carbon dioxide, fossil' in col:   
            C[col] = cf_touse['CO2_IRF'].values
        elif "Carbon dioxide, to soil or biomass stock" in col :   
            C[col] = cf_touse['CO2_IRF'].values * (-1) 
        elif "Carbon dioxide, from soil or biomass stock" in col :   
            C[col] = cf_touse['CO2_IRF'].values 

    elif  "Methane" in col: 
        C[col] = cf_touse['CH4_IRF'].values 
         
    else: 
        C[col] = cf_touse['N2O_IRF'].values 
        
```

```python
C
```

###  new_method2 


Now we want to transform that dataframe to the list of tuples that brightway wants.
- new_method2: 100 new_method for each GWP[t]

- chunk it into like 100 new_method, caz <font color = 'red'> essentially, we have 100 (excl year_0) LCIA, then in calculating lca(), you calculate 100 times for each dGWP, but how can we do it efficiently? </font>


#### note that one stressor mapped to multiple compartment, and we'll use same CF

```python
# for new_method2
data = []  #101 list of tuple for SSP119_MY2020, 101 new_method
for ind in C.index: # to loop through all years 
    irft = []
    for stressor in C.columns:
        all_efs = [i for i in mybio if i['name'] == stressor]
        """ you'll have multiple compartment for one emission """ 
        for ef in all_efs: 
            stressor_code = ef['code']
            #print(f" for {stressor}, for Year{ind}, you have code: {stressor_code} ")
            """ we still relying on 'biosphere3', so name it as it is for LCIA """ 
            irft_i = (('biosphere3', stressor_code), C.loc[ind, stressor])  
            irft.append(irft_i)
    data.append(irft)
```

```python
print(len(data))
print(data[0])
data[100]
```

### write to bw2data.method

```python
for i in range(1, len(data)): 
    #('IPCC 2021 - test SSP119 - year1', 'climate change', 'GWP1')
    name = ('IPCC 2021 - test-SSP119_MY2020 - year' + str(i) , 'climate change', 'dIRF'+str(i) )
    new_method = bw2data.Method(name)
    new_method.register()
    new_method.metadata["unit"] = 'Wperm2perkg'
    new_method.write(data[i])
```
