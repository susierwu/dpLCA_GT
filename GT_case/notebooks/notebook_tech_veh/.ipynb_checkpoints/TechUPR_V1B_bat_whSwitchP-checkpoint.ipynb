#### define tech_switch_P:
- need to know what tech are available for the each v_year, then, diff: tech_Y (battery_replacement_Y, fuel_Y) ...
- users choose one avail tech, or compare all available tech.

```python
bat_tech = agb.newEnumParam(
    name = 'gtruck_BEV_battery_tech',   
    values = ['NMC622', 'LFP', 'LTO', 'NCA', 'LSB'], 
    default = 'NMC622',
    description = "choose between battery tech",
    group = 'g_truck')
```

```python
bat_tech_switch_act_list = []
for prm_name in ssp_scn:
    for year_t in ssp_yr: 
        bg_db_ssp_yr = bg_ei_ + prm_name + "_" + str(year_t)            # backgroun database (BD) to use for SSP[x], Year[t]
        copied_prm_UPR_suffix = "_from_" + prm_name + "_" + str(year_t) # copy UPR from BD to PDB for further use/modification if needed
        newUPR_suffix =  "_wh_" + prm_name + "_" + str(year_t)          # naming on any new UPRs created in PDB

        nmc_final = agb.findActivity("V1B_NMC622 battery_final battery pack_perkWh"  + newUPR_suffix, db_name = PDB_NAME)
        lfp_final = agb.findActivity("V1B_LFP battery_final battery pack_perkWh"  + newUPR_suffix, db_name = PDB_NAME)
        lto_final = agb.findActivity("V1B_LTO battery_final battery pack_perkWh"  + newUPR_suffix, db_name = PDB_NAME)
        nca_final = agb.findActivity("V1B_NCA battery_final battery pack_perkWh"  + newUPR_suffix, db_name = PDB_NAME)
        lsb_final = agb.findActivity("V1B_LSB battery production B3_final battery pack_perkWh"  + newUPR_suffix, db_name = PDB_NAME)
        #print(nmc_final, lsb_final)
        bat_tech_switch_act = agb.newSwitchAct(dbname = PDB_NAME,
                                   name = 'bat_tech_switch_act', 
                                   paramDef = bat_tech, 
                                   acts_dict = {'NMC622' : nmc_final ,
                                                'LFP' : lfp_final, 
                                                'LTO': lto_final, 
                                                'NCA': nca_final, 
                                               'LSB' : lsb_final
                                             })

        bat_tech_switch_act_list.append(bat_tech_switch_act)
```

#### using the switch_act,  creating a battery only URP (not combined with V1A_V1B_init manuf   <font color = 'red'> so that can calculate battery impacts as another module, as bat_replacement will be needed at certain years </font>

```python
batonly_switch_allsspyear = []

for prm_name in ssp_scn:
    for year_t in ssp_yr: 
        bg_db_ssp_yr = bg_ei_ + prm_name + "_" + str(year_t)            # backgroun database (BD) to use for SSP[x], Year[t]
        copied_prm_UPR_suffix = "_from_" + prm_name + "_" + str(year_t) # copy UPR from BD to PDB for further use/modification if needed
        newUPR_suffix =  "_wh_" + prm_name + "_" + str(year_t)          # naming on any new UPRs created in PDB

        #  final 1kWh bat_final_whbatswitch
        bat_final_whbatswitch = agb.newActivity(
                db_name = PDB_NAME,
                name="V1B_bat_g-truck BEV_wh_switch_bat" + newUPR_suffix,
                code = "V1B_bat_g-truck_BEV_wh_switch_bat" + newUPR_suffix,
                unit="bat_for_one_truck",
                exchanges={
                    # vehicle wo battery
                    # agb.findActivity("V1A_g-truck BEV Vehicle Production_" + newUPR_suffix, db_name = PDB_NAME): 1,
                    # switch between diff. bat:
                    agb.findActivity("V1B_NMC622 battery_final battery pack_perkWh" + newUPR_suffix, db_name = PDB_NAME) : dict(amount= battery_total_capacity_for_gtruck, input= bat_tech_switch_act),  
                    agb.findActivity("V1B_LFP battery_final battery pack_perkWh" + newUPR_suffix, db_name = PDB_NAME) : dict(amount= battery_total_capacity_for_gtruck, input= bat_tech_switch_act),  
                    agb.findActivity("V1B_LTO battery_final battery pack_perkWh" + newUPR_suffix, db_name = PDB_NAME) : dict(amount= battery_total_capacity_for_gtruck, input= bat_tech_switch_act),  
                    agb.findActivity("V1B_NCA battery_final battery pack_perkWh" + newUPR_suffix, db_name = PDB_NAME) : dict(amount= battery_total_capacity_for_gtruck, input= bat_tech_switch_act),  
                    agb.findActivity("V1B_LSB battery production B3_final battery pack_perkWh" + newUPR_suffix, db_name = PDB_NAME) : dict(amount= battery_total_capacity_for_gtruck, input= bat_tech_switch_act), 
                     },
            )
        
        for exc in bat_final_whbatswitch.getExchange(name = "*", single=False): 
            exc['type']='technosphere'
            exc.save()
        bat_final_whbatswitch.setOutputAmount(1)
        
        agb.printAct(bat_final_whbatswitch)
        print(f"end of creating a new UPR for desired kWh capacity for g-truck with switch_P for SSP {prm_name} and year {year_t}")
        
        #adding to list: 
        batonly_switch_allsspyear.append(bat_final_whbatswitch)
```

```python
ssp_remind_Pname_map = {
    "SSP1-PkBudg500" : "ssp119", "SSP1-PkBudg1150": "ssp126", "SSP2-Base": "ssp245", "SSP5-Base": "ssp585"
}

tech_todolca_list, tech_ssp_list, tech_year_list, tech_score_list = [],[],[],[]
for gtruck_final_whSwitchbat in gtruck_finalwhbat_switch_allsspyear:
    # need a matched YearlyParams for each SSPx:
    str1 = gtruck_final_whSwitchbat.as_dict()["name"]
    str2 = str1.split("remind_")[1]
    year_to_get_inP = int( str2.split("_")[-1] )
    ssp_remind = str2.split("_")[0] 
    ssp_to_get_inP = ssp_remind_Pname_map.get(ssp_remind)
    #print(f"get available tech lsit for SSP {ssp_to_get_inP} year {year_to_get_inP}")
    
    all_app_bat = []  
    for bat_x in ["LIB_NMC622", 'LIB_LFP',  'LIB_LTO', 'LIB_NCA',  "LSB"]:  
        bat_avail = data2.sel(Tech = bat_x, powertrain = "BEV", Module = "V1B_bat", v_type = "poubelle",  
                              SSP = ssp_to_get_inP, year = year_to_get_inP)["tech_avail"].values
        tech_todolca_list.append(bat_x)
        tech_ssp_list.append(ssp_to_get_inP)
        tech_year_list.append(year_to_get_inP)
        
        if np.isnan(bat_avail).all(): 
            print(f"not available for {bat_x} per SSP {ssp_to_get_inP} year {year_to_get_inP}")
            tech_score_list.append(0)
        else: 
            tech_score_list.append(1)
            
            all_app_bat.append(bat_x)
            #print(all_app_bat)
  
all_tech_todolca = pd.DataFrame(zip(tech_todolca_list,tech_ssp_list, tech_year_list, tech_score_list), 
                               columns = ["tech","ssp","year","score"])

all_tech_todolca
```
