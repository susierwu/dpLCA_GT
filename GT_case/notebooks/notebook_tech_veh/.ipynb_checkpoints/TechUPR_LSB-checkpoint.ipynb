### A.0_parameters for LSB (FU per kWh), sheet"V1B_bat_LSB_perkWh", P group "battery": 

```python
#row9, 295.4 Wh/kg (potentially could reach 500 Wh/kg per Benveniste2022;   per Barker2022_SI: "in specific energy of 813.28 Whâ‹…kg-1 of an optimized LiS-ASSB[Ge] configuration on cell level. )
LSB_battery_cell_density = newFloatParam(
        name="LSB_battery_cell_density",
        default=0.5,
        min=0.2954,
        max=0.73,
        distrib=stats.DistributionType.TRIANGLE,
        group="battery",
        unit="kWh/kg",
    )

LSB_cell_neededkg_perkWh = 1/LSB_battery_cell_density  #unit is kg, this is kg cell needed  for 1 kWh battery

LSB_cell_assmb_NG_MJperkg = newFloatParam(
        name="LSB_cell_assmb_NG_MJperkg",
        default= 5.47,
        distrib=stats.DistributionType.LINEAR,
        min=2.88,
        max=5.47,
        group="battery",
        description="calc from Nature2023", #7.84 3.09 2.88
        unit="MJ",
    )

LSB_cell_assmb_elec_kWhperkg = newFloatParam(
        name="LSB_cell_assmb_elec_kWhperkg",
        default= 1.38,
        distrib=stats.DistributionType.LINEAR,
        min=0.65,
        max=1.38,
        group="battery",
        description="calc from Nature2023", # 1.78 0.97 0.65
        unit="kWh",
    )
```

### B.1  read in LSB parameter under different SSPx in the calc

- using the UPR names created in PDB, to map with the corresponding SSPx' yearlyP to get
- as the URP names has different naming "_remind_SSP1-PkBudg1150" from the Parameter file, need a mapping dict
- once YearlyP under a SSPx extracted, run the agb.lca() calculation 

```python
#final output to be appended
lsb_lcia_dict = {}
GWP1_list, GWP2_list, SSP_list, Year_list = [],[],[],[]

for lsb_final in LSB_final_allsspyear:
    # need a matched YearlyParams for each SSPx:
    str1 = lsb_final.as_dict()["name"]
    str2 = str1.split("remind_")[1]
    year_to_get_inP = int(str2.split("_")[-1] )
    ssp_remind = str2.split("_")[0] 
    ssp_to_get_inP = ssp_remind_Pname_map.get(ssp_remind)
    
    yearlyParams = loadYearlyParams(PARAMS_FILE = p_file, s_name = "V1B_bat_LSB_perkWh", years = [year_to_get_inP], SSP = ssp_to_get_inP) 
    #print(f"for SSP {ssp_to_get_inP}, year {year_to_get_inP}, P:  {yearlyParams}" )
    
    df = agb.multiLCAAlgebric(models=[
                (lsb_final, 1)], 
                methods = chosen_methods, 
                **yearlyParams
                )
            
    print(f"end of running LCA calc for SSP {ssp_to_get_inP} and year {year_to_get_inP} for {lsb_final} ")
    #display(df)
    
    # prepare for final DF: 
    GWP1 = df.iloc[:,0].values[0]
    GWP2 = df.iloc[:,1].values[0]
    GWP1_list.append(GWP1)
    GWP2_list.append(GWP2)
    SSP_list.append(ssp_to_get_inP) 
    Year_list.append(year_to_get_inP)
    
    # assign df as value to each UPR name (used as key)
    lsb_lcia_dict[str1] = df
```

```python
lcia_lsb = pd.DataFrame(zip(SSP_list, Year_list, GWP1_list, GWP2_list), columns = ["SSP","Year","GWPkgCO2e_perkWhLSB", "prm_GWPkgCO2e_perkWhLSB"])
lcia_lsb
```
